# Test the Effects of Rarefaction Depth on Relative Alpha Diversity

This notebook uses GCMP skeletal samples with high sampling depth to test whether choosing a rarefaction depth of 1000 vs. 100,000 sequences per sample would have much effect on which samples appear to have the most biodiverse microbiomes.

## Load alpha diversity data for 1k vs. 100k rarefied GCMP skeleton samples

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

div_df = pd.read_csv("../output/adiv_comparison_table.txt",sep="\t")
div_df = div_df.set_index("sample_id_100k")
div_df
```

## Plot observed features for the same samples at 1k (x-axis) or 100k (y-axis) sequences per sample

```python

x = list(div_df["observed_features_100k"])
y = list(div_df["observed_features_1k"])

plt.scatter(x,y, color = "black", marker = "o")
plt.xlabel("Observed Features \n (100k rarefaction depth)",size="xx-large")
plt.ylabel("Observed Features \n (1k rarefaction depth)",size = "xx-large")
plt.show()

## Fit a regression line
import statsmodels.formula.api as smf
formula = "observed_features_100k ~ observed_features_1k -1"
model = smf.ols(formula,data=div_df)
model_fit = model.fit()
print(model_fit.summary())
```

## Compare *rank* alpha diversity by rarefaction depth (e.g. as in Spearman Correlation)

```python
import statsmodels

## Fit a regression line
import statsmodels.formula.api as smf

div_df_ranked = div_df.rank(axis=0)
formula = "observed_features_100k ~ observed_features_1k -1"
model = smf.ols(formula,data=div_df_ranked)
model_fit = model.fit()

#Set up a Figure and Axes object
fig,ax = plt.subplots()



x = list(div_df_ranked["observed_features_100k"])
y = list(div_df_ranked["observed_features_1k"])

#Plot the raw data
ax.scatter(x,y,marker='o',color='black')

#Set x- and y- axis labels
ax.set_xlabel("Rank Observed Features \n(100k Rarefaction Depth)",size="xx-large")
ax.set_ylabel("Rank Observed Features \n(1k Rarefaction Depth)",size="xx-large")

#Print summary results
print(model_fit.summary())

#Save the figure 
# dpi controls the resolution in dots per inch (important since journals have minimal dpi requirements)
fig.savefig("effect_of_rarefaction_on_rank_diversity_1k_vs_100k",dpi=300)
```
